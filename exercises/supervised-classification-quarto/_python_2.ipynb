```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# First plot: θ0 = 0, θ1 = θ2 = 1
plt.figure(figsize=(4, 4))
x = np.array([-2, 2])
y = -x
plt.plot(x, y, 'k--', label=r'$	heta_0 = 0, 	heta_1 = 	heta_2 = 1$')
plt.scatter([0, 1], [0, -1], c='black', marker='x')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('θ_0 = 0, θ_1 = θ_2 = 1')
plt.grid(True)
plt.show()

# Second plot: θ0 = 1, θ1 = θ2 = 1
plt.figure(figsize=(4, 4))
y = -x - 1
plt.plot(x, y, 'r--', label=r'$	heta_0 = 1, 	heta_1 = 	heta_2 = 1$')
plt.scatter([0, 1], [-1, -2], c='red', marker='x')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('θ_0 = 1, θ_1 = θ_2 = 1')
plt.grid(True)
plt.show()

# Third plot: θ0 = 0, θ1 = 1, θ2 = 2
plt.figure(figsize=(4, 4))
y = -0.5 * x
plt.plot(x, y, 'b--', label=r'$	heta_0 = 0, 	heta_1 = 1, 	heta_2 = 2$')
plt.scatter([0, 1], [0, -0.5], c='blue', marker='x')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('θ_0 = 0, θ_1 = 1, θ_2 = 2')
plt.grid(True)
plt.show()
```
